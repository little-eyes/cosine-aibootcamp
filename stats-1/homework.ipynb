```python
# rolling a dice
import numpy as np
import seaborn as sns

def roll_dice():
    return np.random.randint(1, 7)

s = np.array([roll_dice() for i in range(10000)])
ax = sns.distplot(s)
```

```python
# calculate expectation of the dice rolling
np.mean(s)
```

```python
# calculate variance of the dice rolling
np.var(s)
```

```python
# P(A = 2)
def p(s, target):
    return s[s == target].shape[0] / s.shape[0]

p(s, 2)
```

```python
# P(A = 2, B = 3)
p(s, 2) * p(s, 3)
```

```python
# empirical validation
count = sum([1 if s[i] == 2 and s[i+1] == 3 else 0 for i in range(len(s) - 1)])
count / len(s)
```

```python
# Draw samples from Bernoulli distribution
n, p = 10, 0.5 # number of trials, probability of each trial
results = np.random.binomial(n, p, 1000)
# result of flipping a coin 10 times, tested 1000 times.
ax = sns.distplot(results)
```

```python
# Letâ€™s do 20,000 trials of the model, and count the number that generate zero positive results.
# answer = 0.38885, or 38%.
sum(np.random.binomial(9, 0.1, 20000) == 0)/20000.
```

```python
# Draw samples from standard normal distribution
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
ax = sns.distplot(s)
```

```python
# Verify the mean and the variance
abs(mu - np.mean(s)) < 0.01
```

```python
abs(sigma - np.std(s, ddof=1)) < 0.01
```

```python
# Draw samples from poisson distribution
lam, k = 5, 10000 # lambda and k
s = np.random.poisson(lam, k)
ax = sns.distplot(s)
```
